```python
nombre = "Ivan Dominguez Lugo"
```

```python
import string 

# Código para cifrado César
def cifrado_cesar(texto, k, en_espanol = False):
    alfabeto = list(string.ascii_lowercase)

    if en_espanol: 
        alfabeto.insert(14, "ñ")
    
    texto_cifrado = ""
    
    for caracter in texto:        
        if caracter.lower() in alfabeto:
            numero_de_letra = alfabeto.index(caracter.lower())
            if caracter.islower():                
                texto_cifrado += alfabeto[(numero_de_letra + k) % len(alfabeto)]        
            else:
                texto_cifrado += alfabeto[(numero_de_letra + k) % len(alfabeto)].upper()
        else:
            texto_cifrado += caracter
    
    return texto_cifrado
        

print(cifrado_cesar(nombre, 5, en_espanol=True))
```

```python
# Código para cifrado de sustitución monoalfabética
def sustitucion_monoalfabetica(texto, dict_sustitucion):
    texto_cifrado = ""
    for caracter in texto:
        if caracter.lower() in dict_sustitucion:
            if caracter.islower():
                texto_cifrado += dict_sustitucion[caracter.lower()]                       
            else:
                texto_cifrado += dict_sustitucion[caracter.lower()].upper()
        else:
            texto_cifrado += caracter
    
    return texto_cifrado

sustitucion = {
    'a': 'z',
    'b': 'y',
    'c': 'x',
    'd': 'w',
    'e': 'v',
    'f': 'u',
    'g': 't',
    'h': 's',
    'i': 'r',
    'j': 'q',
    'k': 'p',
    'l': 'o',
    'm': 'ñ',
    'n': 'a',
    'ñ': 'b',
    'o': 'c',
    'p': 'd',
    'q': 'e',
    'r': 'f',
    's': 'g',
    't': 'h',
    'u': 'i',
    'v': 'j',
    'w': 'k',
    'x': 'l',
    'y': 'm',
    'z': 'n'
}

print(sustitucion_monoalfabetica(nombre, sustitucion))
```

```python
llave_vigenere = "examen"

def normalizar(texto):
    return texto.lower().replace(" ", "")

def cifrado_vigenere(texto, llave, en_espanol=False):
    texto = normalizar(texto)
    llave = normalizar(llave)
    
    alfabeto = list(string.ascii_lowercase)
    if en_espanol:
        alfabeto.insert(14, "ñ")

    texto_numeros = []
    llave_numeros = []

    for caracter in texto:
        texto_numeros.append(alfabeto.index(caracter)+1)
    for caracter in llave:
        llave_numeros.append(alfabeto.index(caracter)+1)

    cifrado = ""
    
    for numero in range(len(texto_numeros)):                
        cifrado_num = texto_numeros[numero] + llave_numeros[numero % len(llave_numeros)]
        cifrado_num %= len(alfabeto)
        cifrado += alfabeto[cifrado_num-1]

    return cifrado

print("Texto cifrado sin espacios: {}".format(cifrado_vigenere(nombre, llave_vigenere, en_espanol=True)))
```

```python
# En este código consideré que en llave efectivamente no se repiten los caracteres en la llave y que esta tiene menos de 25 caracteres
llave_playfair = "claves"

def generar_matriz(llave):
    alfabeto = list(string.ascii_lowercase)
    alfabeto.remove('j')        

    llave_separada = [llave[i:i+5] for i in range(0, len(llave), 5)]        

    matriz = []

    for cadena in llave_separada:
        matriz.append([caracter for caracter in cadena])
        

    for caracter in llave: 
        if caracter in alfabeto:
            alfabeto.remove(caracter)    

    while len(alfabeto):
        letra = alfabeto.pop(0)
        if len(matriz[-1]) < 5:
            matriz[-1] += letra
        else:                        
            matriz.append([letra])

    return matriz
    
def devolver_tupla_cifrada(tupla, matriz):
    tupla_separada = [caracter for caracter in tupla]
    posicion_1 = []
    posicion_2 = []

    for i in range(5):
        for j in range(5):
            if matriz[i][j] == tupla_separada[0]:
                posicion_1 = [i, j]
            elif matriz[i][j] == tupla_separada[1]:
                posicion_2 = [i, j]

    tupla_cifrada = ""

    if posicion_1[0] == posicion_2[0]:
        tupla_cifrada += matriz[(posicion_1[0]+1)%5][posicion_1[1]]
        tupla_cifrada += matriz[(posicion_2[0]+1)%5][posicion_2[1]]
    elif posicion_1[1] == posicion_2[1]:
        tupla_cifrada += matriz[posicion_1[0]][(posicion_1[1]+1)%5]
        tupla_cifrada += matriz[posicion_2[0]][(posicion_2[1]+1)%5]
    else:
        tupla_cifrada += matriz[posicion_1[0]][posicion_2[1]]
        tupla_cifrada += matriz[posicion_2[0]][posicion_1[1]]

    return tupla_cifrada

def cifrado_playfair(texto, llave):
    llave = normalizar(llave)
    texto = normalizar(texto)
    texto = texto.replace('ñ', 'n')

    texto_en_tuplas = [texto[0]]        
    texto = texto[1:]

    while len(texto):                    
        if len(texto_en_tuplas[-1]) == 2:            
            texto_en_tuplas.append(texto[0])
            texto = texto[1:]
        else:
            if texto_en_tuplas[-1] == texto[0]:
                texto_en_tuplas[-1] += 'x' # Caracter neutro
            else:
                texto_en_tuplas[-1] += texto[0] 
                texto = texto[1:]       

    if len(texto_en_tuplas[-1]) == 1:
        texto_en_tuplas[-1] += 'x'

    llave_matriz = generar_matriz(llave)

    texto_cifrado = ""    

    for tupla in texto_en_tuplas:
        texto_cifrado += devolver_tupla_cifrada(tupla, llave_matriz)

    return texto_cifrado


print(cifrado_playfair(nombre, llave_playfair))
```

```python
import random
import numpy as np


def generar_llave_hill(texto, en_espanol = False):
    matriz = []    
    tamano_alfabeto = 26
    if en_espanol:
        tamano_alfabeto = 27
    
    texto = normalizar(texto)

    for i in range(len(texto)):
        fila = []
        for j in range(len(texto)):
            fila.append(random.randint(0, tamano_alfabeto-1))
        matriz.append(fila)

    return matriz

def cifrado_hill(texto, llave_cifrado, en_espanol):
    alfabeto = list(string.ascii_lowercase)
    if en_espanol:
        alfabeto.insert(14, "ñ")

    texto = normalizar(texto)
    texto_separado = [caracter for caracter in texto]

    texto_numeros = [alfabeto.index(caracter) for caracter in texto_separado]

    cifrado_matriz = np.dot(texto_numeros, llave_cifrado)

    texto_cifrado = ""

    for caracter in cifrado_matriz:
        texto_cifrado += alfabeto[caracter%len(alfabeto)]

    return texto_cifrado

llave_hill = generar_llave_hill(nombre, True)
print(llave_hill)

print(cifrado_hill(nombre, llave_hill, True))

```
